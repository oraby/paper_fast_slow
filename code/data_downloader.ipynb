# Save location

```python
DELETE_ZIP_AFTER_EXTRACT = True
```

```python
from pathlib import Path
DOWNLOAD_DIR = Path("../data/2p/")
```

# Data download URL

```python
DOWNLOAD_URL = "https://paperfastslowdata.s3-eu-west-1.amazonaws.com/2p_data.zip"
```

# Download and extract data

```python
import zipfile
import requests
import io
from tqdm.auto import tqdm
# Download the zip file
zip_still_downloading_fp = DOWNLOAD_DIR / "2p_data.zip.tmp"
zip_downloaded_fp = DOWNLOAD_DIR / "2p_data.zip"
with open(zip_still_downloading_fp, "wb") as f:
    print("Downloading data...")
    response = requests.get(DOWNLOAD_URL, stream=True)
    total_size_in_bytes= int(response.headers.get('content-length', 0))
    block_size = 4096 #4 Kibibyte
    progress_bar = tqdm(total=total_size_in_bytes, unit='iB', unit_scale=True)
    for data in response.iter_content(block_size):
        progress_bar.update(len(data))
        f.write(data)
    progress_bar.close()

# Rename the file to remove .tmp
zip_still_downloading_fp.rename(zip_downloaded_fp)
# Unzip the file
with zipfile.ZipFile(zip_downloaded_fp, 'r') as zip_ref:
    print("Extracting data...")
    zip_ref.extractall(DOWNLOAD_DIR)
    print("Data extracted to:", DOWNLOAD_DIR)
# Optionally, delete the zip file after extraction
if DELETE_ZIP_AFTER_EXTRACT:
    print("Deleting zip file...")
    zip_downloaded_fp.unlink()
```
